```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url ='https://en.wikipedia.org/wiki/Kabir_Singh'
```

```
# we are sending a request of url and getting response of the request using get()
response=requests.get(url)
response
```

```
response.text
```

```
#Third part libraries - Parsing library - html.parser , lxml
soup=BeautifulSoup(response.text,"lxml")
```

```
soup
```

```
#a stands for anchor tag
#href is hyper refernce for link
for alink in soup.find_all('a'):
  print(alink.get('href'))
  
```

```
#for content in text format in p tag (paragraph)
#use .text to print only tags and not the tags
for ptag in soup.find_all('p'):
  #print(ptag)
  # now second loop for b tag
  for btag in ptag.find_all('b'):
    print(btag.text)
```

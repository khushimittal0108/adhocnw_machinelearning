```
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```
#reading data from a csvfile
csvf=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```
#print the content of csv file
csvf
```

```
#Visualizations
#Histogram
csvf.hist(figsize=(20,20))
```

```
#scatter plot
sb.pairplot(csvf)
```

```
sb.countplot(csvf['CreditScore'])
```

```
#Line graph
plt.xlabel('Row Number')
plt.ylabel('CreditScore')
plt.plot(csvf['RowNumber'],csvf['CreditScore'])
plt.grid(color='green')
plt.show()
```

```
#bar graph
csvf.groupby('Gender').size().plot('pie',shadow=True,label='Gender',explode=[0,0.1],autopct='%d%%')
```

```
csvf.groupby('Geography').size().plot('pie',shadow=True,label='Geography',explode=[0,0.1,0.05],autopct='%d%%')
```

```
#bar graph
plt.bar(csvf['RowNumber'],csvf['CreditScore'],label='Credit Score')
plt.legend()
plt.show()
```

```
plt.bar(csvf['RowNumber'],csvf['Balance'],label='Balance')
plt.legend()
plt.show()
```

```
plt.bar(csvf['RowNumber'],csvf['EstimatedSalary'],label='Estimated salary')
plt.legend()
plt.show()
```

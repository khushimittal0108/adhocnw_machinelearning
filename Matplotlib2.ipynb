```
import matplotlib.pyplot as plt
```

```
x=[4,7,8,9,11,5]
y=[13,23,5,7,9,5]
player=['virat','dhoni','vijay','dhawan']
runs=[120,44,77,33]
exp=[0.1,0,0,0.2]
```

```
#Line graph
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.grid(color='green')
plt.show()
```

```
#pie chart
plt.pie(runs,labels=player,explode=exp,shadow=True)
plt.show()
```

```
#scatter plot
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(x,y)
plt.grid(color='red')
plt.show()
```

```
#bar graph
plt.bar(x,y,label='pollution')
plt.bar(player,runs,label='cricketscore')
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as ptt
```

```
x=[2,5,8,12,9,15,20,6]
x1=[10,4,7,13,1,3]
```

```
y=[5,9,15,3,7,9,4,16]
y1=[10,6,3,9,11,2]
```

```
#label the x-axis and y-axis
ptt.xlabel("time")
ptt.ylabel("speed")
#to display grid and assign color of grid
ptt.grid(color='green')
#plot a line graph and label each graph
ptt.plot(x,y,label="car1")
ptt.plot(x1,y1,label="car2")
#to show the labels
ptt.legend()
#set max. and min. limit for x and y axis
ptt.xlim(0,22)
ptt.ylim(0,18)
#plot a bar graph 
ptt.bar(x,y)
ptt.bar(x1,y1)
```

```
import numpy as np
```

```
#for initializing the range of start of a 2-D array
x=np.arange(100,126,1)
```

```
#to print all the possible 2-D arrays
for i in x :
  y=np.arange(i,201,5)        #the end limit is 201 because the last value is excluded
  z=np.resize(y,(8,2))        #we have used resize() because the number of elements may be greater than 2x8 i.e.16
  print(z)
```

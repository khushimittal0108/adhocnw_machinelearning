```
import requests
```

```
#accessing web site from url
webdata=requests.get('https://php.net')
```

```
webdata   #response code of htttp protocol
#where 200 means url is ok
#where 302 means url is redirected
#where 404 means url not found
#where 505 means server is not permitting for web scraping
```

```
#now get actual data of any website /url
htmldata=webdata.text
```

```
htmldata
```

```
#importing bs4
from bs4 import BeautifulSoup
```

```
soup=BeautifulSoup(htmldata,'html5lib')
#                 html_data ,parser
```

```
type(soup)
```

```
#finding all data
clean_data=soup.get_text()
```

```
print(clean_data)
```

```
#saving data permanently
with open('mywebdata.txt','w+') as f:
  f.write(clean_data)
```

```
#reading from a file
f=open('mywebdata.txt','r')
mydata=f.read()
f.close()
```

```
type(mydata)
```

```
#split
mydata.split()   #will  split all str data by \n
```

```
#split 
import time
for i in mydata.split():
  print(i)
  time.sleep(1)
```

```
#split
newdata1=[i for i in mydata.split()]
```

```
newdata1
```

```
# NLTK graphs
import matplotlib.pyplot as plt
%matplotlib inline 
#incase matplotlib is not supporting by defaults with pandas or nltk
#to replace plt.show() from any platform
```

```
import nltk
```

```
[i for i in dir(nltk) if 'Freq' in i]   # this will plot data for
```

```
nlpdata=nltk.FreqDist(newdata1)
```

```
nlpdata.plot(20) #top  20 words
```

```
#we can remove stopwords from newdata1 then plot the graph
from nltk.corpus import stopwords
```

```
#removing stopwords
removedata=[i for i in newdata1 if i.lower() not in stopwords.words('english')]
```

```
removedata
```

```
#now apply NLTK FreqDist
nlpremove=nltk.FreqDist(removedata)
```

```
nlpremove.plot(30)
```

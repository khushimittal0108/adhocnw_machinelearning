```
#for using camera
import cv2
#to use imshow in colab we use this
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```
#checking version
cv2.__version__
```

```
#images reading
#image name , image property
#1 is for image in same color channel --GRB or RGB
img=cv2.imread("agnes.png",1)
```

```
#to display image using matplotlib
plt.imshow(img)
```

```
#to display original image
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```
#display image using cv2 imshow
cv2_imshow(img)
# wait for window to close
#mili/micro seconds
#cv2.waitkey(0)
```

```
#split into BGR
x,y,z=cv2.split(img)
cv2_imshow(x)
```

```
cv2_imshow(y)
```

```

cv2_imshow(z)
```

```
#to draw a line in image
cv2.line(img,(0,0),(200,300),(0,0,255),3)
# to draw a rectangle in image
cv2.rectangle(img,(50,50),(100,110),(255,0,0),3)
cv2_imshow(img)
```

```
# to create a block of grey image 
#if we use 255 in place of 200 we get white colored image
myimg=np.full((512,512),200)
cv2_imshow(myimg)
```

```
# to create black image
myimg1=np.zeros((512,512))
cv2_imshow(myimg1)
```

```
#image opened in channel 0 ---no color channel -- black white img---grey
img2=cv2.imread("agnes.png",0)
cv2_imshow(img2)
```

```
# -1 maintain transparency
img3=cv2.imread("agnes.png",-1)
cv2_imshow(img3)
```
